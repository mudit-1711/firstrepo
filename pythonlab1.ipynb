## Basic Output Statements in Python

### Print "Hello World"

Using the `print()` function to display a simple text message.

```python
print("Hello World")
```

### Print Multiple Strings in One `print()`

Passing multiple values separated by commas. They are printed with a space between them by default.

```python
print("Hello", "World", "IT")
```

### Print Multiple Lines

Each standard `print()` statement prints on a new line.

```python
print("Hello")
print("World")
print("Python")
```

### Using `\n` (New Line)

The `\n` character can be used to insert a line break inside a single string.

```python
print("Hello\nWorld\nPython")
```

### Using `sep`

The `sep` parameter specifies a custom string to separate the arguments instead of the default space.

```python
print("2026", "07", "12", sep="-")
```

### Using `end`

The `end` parameter changes what gets printed at the very end of the statement (default is a newline).

```python
print("Hello", end=" ")
print("World")
```

### Using `sep` and `end`

Combining both parameters to format both the separator and the ending behavior.

```python
print("Python", "Lab", sep=" | ", end=" : ")
print("Experiment 1")
```
